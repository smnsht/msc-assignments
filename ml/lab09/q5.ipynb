```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots, cm
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

### (a)

```python
rng = np.random.default_rng(5)
x1 = rng.uniform(size=500)- 0.5
x2 = rng.uniform(size=500)- 0.5
y = x1**2- x2**2 > 0
```

```python
y
```

### (b)

```python
fig, ax = subplots(figsize=(8, 8))
ax.scatter(x1, x2, c=y, cmap=cm.coolwarm);
```

### (c)

```python
X = pd.DataFrame({'x1': x1, 'x2': x2})
clf = LogisticRegression(random_state=0).fit(X, y)
```

### (d)

```python
y_predicted = clf.predict(X)
y_predicted
```

```python
fig, ax = subplots(figsize=(8, 8))
ax.scatter(x1, x2, c=y_predicted, cmap=cm.coolwarm);
```

### (e)

```python
X_non_linear = pd.DataFrame({'x1': x1, 'x1^2': x1**2, 'x2': x2, 'x2^2': x2**2})
X_non_linear
```

```python
clf_non_linear = LogisticRegression(random_state=0).fit(X_non_linear, y)
```

### (f)

```python
y_predicted_non_linear = clf_non_linear.predict(X_non_linear)
fig, ax = subplots(figsize=(8, 8))
ax.scatter(x1, x2, c=y_predicted_non_linear, cmap=cm.coolwarm);
```

### (g)

```python
svm_linear = SVC(C=10, kernel='linear')
svm_linear.fit(X, y)
y_predicted_svm_linear = svm_linear.predict(X)

fig, ax = subplots(figsize=(8, 8))
ax.scatter(x1, x2, c=y_predicted_svm_linear, cmap=cm.coolwarm);
```

### (h)

```python
svm_poly = SVC(C=10, kernel='poly', degree=2)
svm_poly.fit(X, y)
y_predicted_poly = svm_poly.predict(X)

fig, ax = subplots(figsize=(8, 8))
ax.scatter(x1, x2, c=y_predicted_poly, cmap=cm.coolwarm);
```

### (i)

We can also obtain a non-linear decision boundary by performing logistic regression using non-linear transformations of the features.
