```python
from sklearn.model_selection import (train_test_split)
from sklearn.linear_model import LogisticRegression

from ISLP import load_data
from ISLP.models import (ModelSpec as MS)
import numpy as np
```

```python
Default = load_data('Default')
Default
```

```python
# (a)

def a():
    ms = MS(['balance',	'income'])
    X = ms.fit_transform(Default)
    Y = (Default['default'] == 'Yes')

    clf = LogisticRegression(random_state=0).fit(X, Y)
    return clf

a()
```

```python
# (b)


def b(test_size=0.2):
    # split the sample set into a training set and a validation set.
    Default_train, Default_valid = train_test_split(Default, test_size=test_size, random_state=0)


    # fit a multiple logistic regression model using only the training observations.
    ms = MS(['balance',	'income'])
    x_train = ms.fit_transform(Default_train)
    y_train = (Default_train['default'] == 'Yes')        
    clf = LogisticRegression(random_state=0).fit(x_train, y_train)

    x_valid = ms.fit_transform(Default_valid)
    y_valid = (Default_valid['default'] == 'Yes')        

    # Obtain a prediction of default status for each individual in the validation set
    predicted = clf.predict(x_valid)

    #  Compute the validation set error,
    return np.mean(predicted != y_valid)    
    
b()
```

```python
# (c)

b(0.2), b(0.3), b(0.5)

# error seems to be near 27%
```

```python
# (d)

def d(test_size=0.2):
    Default_train, Default_valid = train_test_split(Default, test_size=test_size, random_state=0)
    
    
    ms = MS(['balance',	'income', 'student'])    
    x_train = ms.fit_transform(Default_train)        
    y_train = (Default_train['default'] == 'Yes')               
    clf = LogisticRegression(random_state=0).fit(x_train, y_train)
    x_valid = ms.fit_transform(Default_valid)
    y_valid = (Default_valid['default'] == 'Yes')        

    
    predicted = clf.predict(x_valid)
    
    
    return np.mean(predicted != y_valid)    
    
d(0.2), d(0.3), d(0.5)

# (0.0285, 0.029333333333333333, 0.0288)
# new error is approx. 2.9%
```
