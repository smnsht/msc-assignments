```python
import pandas as pd
from matplotlib.pyplot import subplots
```

```python
boston = pd.read_csv('./Boston.csv').dropna();
boston

# (b)
#
# 506 rows, 14 columns
```

```python
# (c)

fig, axes = subplots(ncols=3, figsize=(16, 4))

x1 = boston['age']
y1 = boston['crim']
axes[0].set_xlabel("proportion of owner-occupied units built prior to 1940")
axes[0].set_ylabel("crime rate")
axes[0].set_title("old homes - more crime")
axes[0].scatter(x1, y1, marker='o');

x2 = boston['dis']
y2 = boston['crim']
axes[1].set_xlabel("distances to employment")
axes[1].set_ylabel("crime rate")
axes[1].set_title("close to work - more crime")
axes[1].scatter(x2, y2, marker='o');

x3 = boston['tax']
y3 = boston['crim']
axes[2].set_xlabel("property-tax rate")
axes[2].set_ylabel("crime rate")
axes[2].set_title("hight tax - more crime")
axes[2].scatter(x3, y3, marker='o');
```

(d)

1 - old homes - more crime
2 - close to work - more crime
3 - hight property tax - more crime

```python
# (e)

fig, ax = subplots(figsize=(8, 8))

x1 = boston['ptratio']
y1 = boston['crim']
ax.set_xlabel("pupil-teacher ratio by")
ax.set_ylabel("crime rate")
ax.set_title("Crime rate dependency of pupil-teacher ratios")
ax.scatter(x1, y1, marker='o');

# less teachers - more crime
```

```python
# (i)
#In this data set, how many of the suburbs average more than seven rooms per dwelling?

over7 = len(boston[boston['rm'] > 7])
over8 = len(boston[boston['rm'] > 8]);

format = 'more the 7: {0}, more then 8: {1}'.format(over7, over8)
print(format)

# more the 7: 64, more then 8: 13
# relatively lower crime
```
